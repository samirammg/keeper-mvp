```python
file_path =  "data/0_sim_contract.xlsx"
df_cnt =  pd.read_excel(file_path)
df_cnt.shape
```

```python
fe=['account_id', 'contract_id', 'contract_start_date', 'contract_end_date','number_contract',  'industry', 'business_size', 'annual_revenue', 'has_product_1', 'has_product_2', 'has_product_3','has_product_4', 'has_product_5', 'contract_acv', 'per_product', 'pre_contract_value_change',
 'pre_contract_value_change_pct', 'first_contract_start_date','age_at_contract', 'contract_duration', 'num_pre_contract','first_contract_value', 'first_contract_value_change',
 'first_contract_value_change_pct', 'acv_label', 'clv_contract', 'clv_contract_with_churn', 'customer_lifespan_contract']

ex=['is_chr','has_next_contract', 'next_contract_value_change', 'next_contract_value_change_category', 'group', 'group_des', 'ischr',
       'isnew', 'isup', 'isdown', 'issame']

aa=fe+['ischr']
```

```python
data=df_cnt[aa]
```

```python
data[data['account_id']=='account_599']
```

```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler, LabelEncoder
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score, classification_report
import matplotlib.pyplot as plt

# Load the dataset
# data = pd.read_csv('your_dataset.csv')  # Replace with your dataset

# Fill missing values if any (using mean for continuous features and mode for categorical features)
data.fillna(0, inplace=True)

# Handle categorical variables (encode 'industry' and 'acv_label')
label_encoder = LabelEncoder()
data['industry'] = label_encoder.fit_transform(data['industry'])
data['acv_label'] = label_encoder.fit_transform(data['acv_label'])

# One-hot encode 'industry' and 'business_size'
data_encoded = pd.get_dummies(data['industry'], prefix='industry')
data = pd.concat([data, data_encoded], axis=1)

data_encoded = pd.get_dummies(data['business_size'], prefix='business_size')
data = pd.concat([data, data_encoded], axis=1)

# Select features and target
X = data.drop(columns=['ischr', 'contract_id', 'account_id', 'industry', 'business_size'])  # Drop target and ID columns
y = data['ischr']  # Target column

# Identify numeric columns for scaling
numeric_cols = X.select_dtypes(include=['int64', 'float64']).columns

# Split data into train and test sets
X_train, X_test, y_train, y_test = train_test_split(X[numeric_cols], y, test_size=0.2, random_state=42)

# Standardize the features
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)

# Initialize and train a Random Forest Classifier
model = RandomForestClassifier(random_state=42)
model.fit(X_train_scaled, y_train)

# Make predictions on the test set
y_pred = model.predict(X_test_scaled)

# Evaluate the model
accuracy = accuracy_score(y_test, y_pred)
class_report = classification_report(y_test, y_pred)

# Display the evaluation metrics
print(f"Accuracy: {accuracy:.2f}")
print(f"Classification Report:\n{class_report}")

# Get feature importance
feature_importance = model.feature_importances_

# Create a DataFrame to view the feature importances
feature_importance_df = pd.DataFrame({
    'Feature': X_train.columns,
    'Importance': feature_importance
})

# Sort the features by importance
feature_importance_df = feature_importance_df.sort_values(by='Importance', ascending=False)

# Display the feature importances
print(feature_importance_df)

# Plot feature importance
plt.figure(figsize=(10, 6))
plt.barh(feature_importance_df['Feature'], feature_importance_df['Importance'])
plt.xlabel('Feature Importance')
plt.title('Feature Importance of Random Forest Model')
plt.gca().invert_yaxis()  # To display the most important features at the top
plt.show()
```

```python
import pandas as pd
from sklearn.preprocessing import StandardScaler

# Assuming 'X' is your feature DataFrame
# Identify numeric columns for scaling
numeric_cols = X.select_dtypes(include=['int64', 'float64']).columns

# Standardize only the numeric features
scaler = StandardScaler()
X_scaled_numeric = scaler.fit_transform(X[numeric_cols])

# Convert the scaled numeric data back to a DataFrame
X_scaled_numeric_df = pd.DataFrame(X_scaled_numeric, columns=numeric_cols, index=X.index)

# Recombine scaled numeric columns with non-numeric columns (if needed)
X_scaled = pd.concat([X_scaled_numeric_df, X.drop(columns=numeric_cols)], axis=1)

# Get predicted values (y_pred) for all rows in the data
y_pred = model.predict(X_scaled[numeric_cols])  # Use only numeric columns for prediction
y_pred_proba = model.predict_proba(X_scaled[numeric_cols])[:, 1]  # Use only numeric columns for prediction

# Add predictions to the original DataFrame
data['score_y_c'] = y_pred_proba.tolist()
data['score_y'] = y_pred.tolist()
```

```python
import pandas as pd
import xgboost as xgb
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import accuracy_score, classification_report

# Assuming 'df' is your DataFrame
# Select features and target
X = df[['FOM','contract_id', 'MOC', 'POC','usage_score','support_score','activity_score', 'score_y_c']]
y = df['ischr']

# Optional: If 'contract_id' is categorical, you might want to drop it or encode it, as it might not be useful for prediction
X = X.drop(columns=['contract_id','FOM'])  # Dropping 'contract_id' assuming it's not a useful predictor

# Fill missing values if any (using mean for continuous features)
X.fillna(X.mean(), inplace=True)

# Standardize the features
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# Split data into train and test sets
X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, test_size=0.2, random_state=42)

# Initialize and train an XGBoost classifier
model = xgb.XGBClassifier(use_label_encoder=False, eval_metric='logloss', random_state=42)
model.fit(X_train, y_train)

# Make predictions on the test set
y_pred = model.predict(X_test)

# Evaluate the model
accuracy = accuracy_score(y_test, y_pred)
class_report = classification_report(y_test, y_pred)

# Display the evaluation metrics
print(f"Accuracy: {accuracy:.2f}")
print(f"Classification Report:\n{class_report}")

# Optionally: Predict probabilities for the positive class
y_pred_proba = model.predict_proba(X_test)[:, 1]
print("Predicted probabilities (scores) for the first few test samples:")
print(y_pred_proba[:10])
```
