```python
import pandas as pd

# ------------------------------------------
# 1. Load Data
# ------------------------------------------
df = pd.read_csv("data/0_sim_act_monthly_features_agg.csv")
df_activity_score = pd.read_csv("data/1_score_activity.csv")

# ------------------------------------------
# 2. Define activity type and topic columns
# ------------------------------------------
activity_types = [
    'invitation_count', 'training_count', 'checkin_count', 'onboarding_count',
    'renewal_pricing_count', 'inperson_count', 'demo_count',
    'gift_count', 'document_count'
]

# ------------------------------------------
# 3. Aggregate Activity Features by FOM
# ------------------------------------------
agg_dict = {
    'total_activities': ['sum', 'median'],
    'meeting_count': ['sum', 'median'],
    'email_count': ['sum', 'median'],
    'postal_count': ['sum', 'median'],
    'not_attend_count': ['sum', 'median'],
    'cancel_count': ['sum', 'median'],
    'attend_count': 'sum',
    'total_duration': 'sum',
    'average_duration': 'median',
    'high_priority_count': 'sum',
}

for col in activity_types:
    if col in df.columns:
        agg_dict[col] = 'sum'

grouped = df.groupby("FOM").agg(agg_dict)

# High priority % of total
grouped[('high_priority_pct', '')] = (
    grouped[('high_priority_count', 'sum')] /
    grouped[('total_activities', 'sum')]
) * 100

# Flatten column names
grouped = grouped.drop(columns=['high_priority_count'])
grouped.columns = ['_'.join(col).strip('_') for col in grouped.columns]
grouped = grouped.sort_index()

# ------------------------------------------
# 4. Cancel & Not Attend Percentages
# ------------------------------------------
status_totals = df.groupby("FOM")[['attend_count', 'cancel_count', 'not_attend_count']].sum()
status_totals["cancel_pct"] = status_totals["cancel_count"] / status_totals.sum(axis=1) * 100
status_totals["not_attend_pct"] = status_totals["not_attend_count"] / status_totals.sum(axis=1) * 100

# ------------------------------------------
# 5. Combine + Calculate MoM % Change (Safe)
# ------------------------------------------
dashboard_with_pct = grouped.merge(status_totals[["cancel_pct", "not_attend_pct"]], left_index=True, right_index=True)

# Calculate MoM % Change
mom_change = dashboard_with_pct.pct_change(periods=1) * 100
mom_change.columns = [col + "_mom_pct_change" for col in mom_change.columns]

# Combine with main metrics
dashboard_final = pd.concat([dashboard_with_pct, mom_change], axis=1)

# Clean up NaN and Inf values
dashboard_final = dashboard_final.replace([float("inf"), float("-inf")], pd.NA)
dashboard_final = dashboard_final.fillna(0)

# ------------------------------------------
# 6. Top 3 Topics per FOM
# ------------------------------------------
existing_topics = [col for col in activity_types if col in df.columns]
topic_sums = df.groupby("FOM")[existing_topics].sum()



# Define function to get top 3
def get_top_3_topics(row):
    top_topics = row.sort_values(ascending=False).head(3).index.tolist()
    top_topics_cleaned = [t.replace('_count', '') for t in top_topics]
    return ', '.join(top_topics_cleaned)

topic_sums["top_3_topics"] = topic_sums.apply(get_top_3_topics, axis=1)
df_final = dashboard_final.merge(topic_sums[["top_3_topics"]], left_index=True, right_index=True)

# ------------------------------------------
# 7. Activity Score Features
# ------------------------------------------
df_activity_score['FOM'] = pd.to_datetime(df_activity_score['FOM'])

# Score percentiles
score_percentiles = df_activity_score.groupby('FOM')['activity_score'].quantile([0.5, 0.9, 1.0]).unstack()
score_percentiles.columns = ['activity_score_p50', 'activity_score_p90', 'activity_score_p100']

# Label counts and percentages
label_counts = df_activity_score.groupby(['FOM', 'activity_score_label']).size().unstack(fill_value=0)
label_counts.columns = [f"label_count_{col.lower()}" for col in label_counts.columns]

label_percentages = label_counts.div(label_counts.sum(axis=1), axis=0) * 100
label_percentages.columns = [f"label_pct_{col.split('_')[-1]}" for col in label_percentages.columns]

# Combine score features
activity_score_features = score_percentiles.join(label_counts).join(label_percentages)

# ------------------------------------------
# 8. Merge All Features
# ------------------------------------------
df_final.index = pd.to_datetime(df_final.index)
activity_score_features.index = pd.to_datetime(activity_score_features.index)

final_df = df_final.merge(activity_score_features, on='FOM', how='left')

# ------------------------------------------
# 9. Final Touches
# ------------------------------------------
final_df = final_df.round(2)
final_df.to_csv("data/5_activity_dashboard.csv", index=False)

# Confirmation
print("Final shape:", final_df.shape)
```

```python
import pandas as pd

# Load data
df = pd.read_csv("data/0_sim_act_monthly_features_agg.csv")

# Define activity types and topic columns
activity_types = [
    'invitation_count', 'training_count', 'checkin_count', 'onboarding_count',
    'renewal_pricing_count', 'inperson_count', 'demo_count',
    'gift_count', 'document_count'
]

# Create aggregation dictionary
agg_dict = {
    'total_activities': ['sum', 'median'],
    'meeting_count': ['sum', 'median'],
    'email_count': ['sum', 'median'],
    'postal_count': ['sum', 'median'],
    'total_duration': 'sum',
    'average_duration': 'median',
    'high_priority_count': 'sum',
}

# Add each activity type with sum only
for col in activity_types:
    if col in df.columns:  # Only include existing columns
        agg_dict[col] = 'sum'

# Group by FOM and aggregate
grouped = df.groupby("FOM").agg(agg_dict)

# Add high priority % of total
grouped[('high_priority_pct', '')] = (
    grouped[('high_priority_count', 'sum')] /
    grouped[('total_activities', 'sum')]
) * 100

# Drop high_priority_count column to clean
grouped = grouped.drop(columns=['high_priority_count'])

# Flatten column names
grouped.columns = ['_'.join(col).strip('_') for col in grouped.columns]

# Sort FOM for correct MoM calculation
grouped = grouped.sort_index()

# Calculate Month-over-Month % change
mom_change = grouped.pct_change(periods=1) * 100
mom_change.columns = [col + "_mom_pct_change" for col in mom_change.columns]

# Combine original metrics with MoM changes
dashboard_final = pd.concat([grouped, mom_change], axis=1)

# --- Add Top 3 Most Frequent Topics per FOM ---

# Only use existing topic columns
existing_topics = [col for col in activity_types if col in df.columns]

# Group sums of topics per FOM
topic_sums = df.groupby("FOM")[existing_topics].sum()

# Define function to get top 3
def get_top_3_topics(row):
    top_topics = row.sort_values(ascending=False).head(3).index.tolist()
    top_topics_cleaned = [t.replace('_count', '') for t in top_topics]
    return ', '.join(top_topics_cleaned)

# Apply function row-wise
topic_sums["top_3_topics"] = topic_sums.apply(get_top_3_topics, axis=1)

# Merge with main dashboard
df_final = dashboard_final.merge(topic_sums[["top_3_topics"]], left_index=True, right_index=True)


# Optionally: save to CSV
df_final.to_csv("data/5_activity_dashboard.csv", index=False)
```

```python
import pandas as pd
import numpy as np
from datetime import timedelta
from scipy.stats import zscore
import random

# Convert to DataFrame
contract_df = pd.DataFrame(contract_data)
contract_df['contract_start_date'] = pd.to_datetime(contract_df['contract_start_date'])
contract_df['contract_end_date'] = pd.to_datetime(contract_df['contract_end_date'])

# Define contract_status based on ischr, isup, and isrenew
conditions = [
    (contract_df["ischr"] == 1),
    (contract_df["isup"] == 1)
    ]
choices = ["churned", "upsell"]
contract_df["contract_status"] = np.select(conditions, choices, default="unknown")


# Activity types and topics
activity_types = ['Email', 'Meeting', 'Postal']
email_topics = ['training', 'renewal', 'pricing', 'document', 'onboarding', 'checkin', 'invitation']
meeting_topics = ['training', 'renewal', 'onboarding', 'checkin', 'demo', 'invitation', 'inperson']
postal_topics = ['training', 'invitation', 'gift']

# Generate interactions
interactions = []

for _, contract in contract_df.iterrows():
    csm_id = f"csm_{random.randint(1, 5)}"  # Assign a CSM ID
    num_interactions = random.randint(5, 20) if contract['ischr'] == 0 else random.randint(1, 5)  # Fewer interactions for churned accounts
    contract_status=contract['contract_status']
    contract_start_date=contract['contract_start_date']
    for _ in range(num_interactions):
        activity_type = random.choices(activity_types, weights=[0.2, 0.7, 0.1])[0]  # Majority are meetings
        
        if activity_type == 'Email':
            activity_topic = random.choice(email_topics)
            duration = 0
        elif activity_type == 'Meeting':
            activity_topic = random.choice(meeting_topics)
            duration = random.choice([30, 60, 90, 120])
        else:
            activity_topic = random.choice(postal_topics)
            duration = 0
        
        # Generate a random date within the contract period
        date = contract['contract_start_date'] + timedelta(days=random.randint(0, (contract['contract_end_date'] - contract['contract_start_date']).days))
        
        FOM= date.replace(day=1)
        # Assign priority
        priority = random.choice(['high', 'medium', 'low'])
        
        # Assign status and sentiment
        if activity_type == 'Meeting':
            status = random.choices(['attend', 'cancel', 'not_attend'], weights=[0.83, 0.05, 0.12])[0]  # 80% attend, 10% cancel, 10% not_attend
            sentiment_class = random.choices(['negative', 'positive', 'neutral'], weights=[0.25, 0.30, 0.45])[0] if status == 'attend' else None
        else:
            status = None
            sentiment_class = None
        
        interactions.append({
            'date': date,
            'FOM':FOM,
            'account_id': contract['account_id'],
            'contract_id': contract['contract_id'],
            'csm_id': csm_id,
            'activity_type': activity_type,
            'activity_topic': activity_topic,
            'duration (minutes)': duration,
            'priority': priority,
            'status': status,
            'sentiment_class': sentiment_class,
            'contract_status':contract_status,
            'contract_start_date':contract_start_date
        })

# Convert to DataFrame
interaction_df = pd.DataFrame(interactions)
interaction_df['contract_year'] = interaction_df['FOM'].dt.year
```

```python
import pandas as pd
from datetime import datetime

df = interaction_df.copy()

# Convert 'date', 'FOM', and 'contract_year' columns to datetime
df['date'] = pd.to_datetime(df['date'])
df['FOM'] = pd.to_datetime(df['FOM'])
df['contract_year'] = pd.to_datetime(df['contract_year'], format='%Y')

# Initialize a dictionary to store the results
results = []

# Group by 'contract_id' and 'FOM'
grouped = df.groupby(['contract_id', 'FOM'])

# Iterate over each group
for (contract_id, fom), group in grouped:
    # Calculate Month of Contract (MOC)
    start_of_contract = group['contract_start_date'].iloc[0]  # Start of the contract year
    moc = (fom.year - start_of_contract.year) * 12 + (fom.month - start_of_contract.month) + 1
    contract_status=group['contract_status'].iloc[0]
    # Calculate Percentage of Contract Completion (POC)
    poc = moc / 12

    # Initialize a dictionary to store the features for this group
    features = {
        'contract_id': contract_id,
        'FOM': fom,
        'MOC': moc,
        'start_of_contract':start_of_contract,
        'POC': poc,
        'contract_status':contract_status,
        'total_activities': len(group),
        'meeting_count': len(group[group['activity_type'] == 'Meeting']),
        'email_count': len(group[group['activity_type'] == 'Email']),
        'postal_count': len(group[group['activity_type'] == 'Postal']),
        'invitation_count': len(group[group['activity_topic'] == 'invitation']),
        'training_count': len(group[group['activity_topic'] == 'training']),
        'checkin_count': len(group[group['activity_topic'] == 'checkin']),
        'onboarding_count': len(group[group['activity_topic'] == 'onboarding']),
        'renewal_pricing_count': len(group[group['activity_topic'].isin(['renewal', 'pricing'])]),
        'inperson_count': len(group[group['activity_topic'] == 'inperson']),
        'demo_count': len(group[group['activity_topic'] == 'demo']),
        'gift_count': len(group[group['activity_topic'] == 'gift']),
        'document_count': len(group[group['activity_topic'] == 'document']),
        'total_duration': group['duration (minutes)'].sum(),
        'average_duration': group[(group['duration (minutes)'] > 0) & 
                                  (~group['status'].isin(['cancel', 'not_attend']))]['duration (minutes)'].mean(),
        'attend_count': len(group[group['status'] == 'attend']),
        'not_attend_count': len(group[group['status'] == 'not_attend']),
        'cancel_count': len(group[group['status'] == 'cancel']),
        'high_priority_count': len(group[group['priority'] == 'high']),
        'negative_sentiment_count': len(group[group['sentiment_class'] == 'negative'])
    }
    
    # Append the features to the results list
    results.append(features)

# Convert the results to a DataFrame
results_df = pd.DataFrame(results)

# Save the results to a new CSV file
results_df.shape
```

```python
# Sort the DataFrame by 'contract_id' and 'FOM' for cumulative calculations
results_df = results_df.sort_values(by=['contract_id', 'FOM'])

# List of features to calculate incremental values and MoM changes
features_to_aggregate = [
    'total_activities', 'meeting_count', 'email_count', 'postal_count', 
    'invitation_count', 'training_count', 'checkin_count', 'onboarding_count', 
    'renewal_pricing_count', 'inperson_count', 'demo_count', 'gift_count', 
    'document_count', 'total_duration', 'attend_count', 'not_attend_count', 
    'cancel_count', 'high_priority_count', 'negative_sentiment_count'
]

# Calculate incremental values (cumulative sum) for each feature
for feature in features_to_aggregate:
    results_df[f'agg_{feature}'] = results_df.groupby('contract_id')[feature].cumsum()

# Calculate rolling average for 'average_duration'
results_df['rolling_avg_average_duration'] = results_df.groupby('contract_id')['average_duration'].expanding().mean().reset_index(level=0, drop=True)

# Calculate month-over-month (MoM) changes for specified features
mom_features = [
    'total_activities', 'total_duration', 'attend_count', 
    'not_attend_count', 'cancel_count', 'high_priority_count', 
    'negative_sentiment_count'
]

for feature in mom_features:
    results_df[f'mom_{feature}'] = results_df.groupby('contract_id')[feature].diff().fillna(0)

# Save the results to a new CSV file
results_df.shape
```

```python
features= ['agg_total_activities', 'agg_total_duration', 'agg_attend_count', 'agg_not_attend_count', 'agg_cancel_count', 'agg_high_priority_count', 'agg_negative_sentiment_count']
results_df['contract_year'] = results_df['FOM'].dt.year
data=results_df.copy()
# Filter data for ischr == True
data_chr = data[data['contract_status'] == 'churned']

stats_chr = data_chr.groupby(['contract_year'])[features].agg(['min', 'mean', 'median', 'max', 'std']).reset_index()

# Filter data for has_next_contract == True
data_renew = data[data['contract_status'] == 'upsell']
stats_renew = data_renew.groupby(['contract_year'])[features].agg(['min', 'mean', 'median', 'max', 'std']).reset_index()

# Initialize an empty dictionary to store results
stats_dict = {}

# Loop through both stats for 'ischr' and 'has_next_contract' and add to the dictionary
for stat_df, key in zip([stats_chr, stats_renew], ['ischr', 'isup']):
    # Loop through features and stats
    for feature in features:
        for stat in ['min', 'mean', 'median', 'max', 'std']:
            # Create a key combining feature, stat, contract_year, and 'ischr' or 'has_next_contract'
            for contract_year in stat_df['contract_year']:
                # Add each result to the dictionary
                stats_dict[(feature, stat, contract_year, key)] = stat_df.loc[stat_df['contract_year'] == contract_year, (feature, stat)].values[0]
```

```python
#stats_dict
df = pd.DataFrame(
    [(feature, stat, year, category, value) for (feature, stat, year, category), value in stats_dict.items()],
    columns=['Feature', 'Statistic', 'Year', 'Category', 'Value']
)
df.to_csv("data/0_sim_activity_yearly_fuzzy.csv", index=False)
```

```python
for f in features:
    # Iterate through each row in data
    for idx, row in data.iterrows():
        year = row['contract_year'] - 1 if row['contract_year'] > data['contract_year'].min() else row['contract_year']
        
        # Get the median values from stats_dict for 'ischr' and 'isup'
        median_ischr = stats_dict.get((f, 'median', year, 'ischr'), None)
        median_renew = stats_dict.get((f, 'median', year, 'isup'), None)

        # Check if the median values exist, otherwise assign 'normal'
        if median_ischr is not None and row[f] < median_ischr:
            label = 'risky'
        elif median_renew is not None and row[f] > median_renew:
            label = 'healthy'
        elif median_ischr is not None and median_renew is not None and median_ischr <= row[f] <= median_renew:
            label = 'normal'
        else:
            label = 'noun'

        # Assign the label to the row (new column with label for that feature)
        data.at[idx, f + '_label'] = label
        data.at[idx, f + '_median_ischr'] = median_ischr
        data.at[idx, f + '_median_isup'] = median_renew
```

```python
a=[]
for i in  data.columns.tolist():
    if ('label' in i ):
        a.append(i)
aa=a+['ischr','isup']
```

```python
# Define a mapping function to convert categorical values into -1, 0, 1
def map_categories_to_numeric(value):
    if value == 'risky':  # Risky category mapped to -1
        return -1
    elif value == 'healthy':  # Healthy category mapped to 1
        return 1
    else:  # Neutral or normal categories mapped to 0
        return 0

# Apply the mapping function to the categorical columns
categorical_columns=[]
for i in data.columns:
    if '_label' in i :
        categorical_columns.append(i)
        
for col in categorical_columns:
    data[col] = data[col].apply(map_categories_to_numeric)

weights = {}
for col in a:
    # Calculate the correlation of each feature with the target variable 'ischr'
    correlation = data[col].corr(data['ischr'])
    weights[col] = correlation

# Create a DataFrame to display the feature weights
weights_df = pd.DataFrame(list(weights.items()), columns=['Feature', 'Weight'])
weights_df

wl=[]
for i in weights_df['Weight'].tolist():
    #wl.append(i/sum(l))
    wl.append(1/len(a))
data['activity_score']=0
for i,j in zip(a,wl):
    data['activity_score']=data['activity_score']+data[i]*j

data['activity_score_label'] = data['activity_score'].apply(lambda x: 'healthy' if x > 0 else ('normal' if x == 0 else 'risky'))
```
